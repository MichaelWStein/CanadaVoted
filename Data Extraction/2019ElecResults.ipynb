This program creates *.csv files for all electoral resulta. The election results have been retrieved from elections Canada. 

2019 Election Results: https://enr.elections.ca/National.aspx?lang=e 


```python
import pandas as pd
```

```python
results_2019 = pd.read_csv("2019_Election_Results.csv")
print(results_2019.columns)
results_2019.head()
```

```python
#Removing unnecessary columns

results_2019 = results_2019.drop(['Unnamed: 2', 'Unnamed: 4', 'Unnamed: 5',
       'Unnamed: 6', 'Unnamed: 7', 'Unnamed: 9', 'Unnamed: 12', 'Unnamed: 13'], axis=1)

print(results_2019.columns)

results_2019.head()
```

```python
#Renaming Columns
results_2019 = results_2019.rename(columns={'These results were downloaded and refreshed on February  04, 2020 at 12:40 p.m. ET. / Ces résultats ont été téléchargés et rafraichis le 4 février 2020 à 12h40 HE.': "Electoral_District",
       'Unnamed: 1': "Elec_Dist_Name", 'Unnamed: 3': "Result_type", 'Unnamed: 8':"Party", 'Unnamed: 10':"Votes", 'Unnamed: 11':"Votes_%"})
                                   
results_2019.head()
```

```python
#Checking if all electoral districts have validated results
compare1 = results_2019["Electoral_District"].unique()
compare2 = (results_2019.loc[results_2019['Result_type']=='validated'])["Electoral_District"].unique()

print(list(set(compare1)-set(compare2)))

#There are no districts that don't have validated results, so the preliminary results can be eliminated

results_2019 = results_2019.loc[results_2019['Result_type']=='validated']

#Columns

results_2019.head()
```

```python
#Reset the index, remove Result_type and save file as *.csv for future processing

results_2019 = results_2019.drop(['Result_type'], axis=1)
results_2019 = results_2019.reset_index()
results_2019 = results_2019.drop(["index"], axis=1)

results_2019.to_csv('2019_Results_Complete.csv')
results_2019.head()
```

```python
#Need to ensure that Votes and Votes_% are all in numeric

results_2019['Votes']=pd.to_numeric(results_2019['Votes'])
results_2019['Votes_%']=pd.to_numeric(results_2019['Votes_%'])

results_2019.info(verbose=True)
```

```python
#Simplify - Create a dataframe with only the winner

#Create a list of electoral districts
elec_list = results_2019["Electoral_District"].unique()

#Loop through the dataframe. Put the index for the row with the highest percentage in a list

win_list = []

#Make sure the Dataframe is empty
results_simp=pd.DataFrame()

for district in elec_list:
    eval_frame=results_2019.loc[results_2019['Electoral_District']==district]
    result=eval_frame.loc[eval_frame['Votes']==eval_frame['Votes'].max()]
    win_list.append(result.index)

x = len(win_list)

for i in win_list:
    results_simp = results_simp.append(results_2019.iloc[i])

results_simp.head()


```

```python
#Update the dataframe index and add dates to the column names

results_simp = results_simp.reset_index()
results_simp = results_simp.drop(["index"], axis=1)

results_simp = results_simp.rename(columns={'Party': "2019_Party", 'Votes': '2019_Votes', 'Votes_%': '2019_Votes_%'})

dist_name = results_simp['Elec_Dist_Name'].tolist()
dist_num = results_simp['Electoral_District'].tolist()

results_simp.head()
       
                                   

```

```python
#Save the results as csv and pickle for further processing.

results_simp.to_csv('2019_Results.csv')
results_simp.to_pickle("2019_Results.pkl")
```

Uploading the data from statscan through json. Needs to be uploaded with a string and then the first two elements cut off, after that jsonify. This notebook is just for one electoral dsitrict, basically as a 'sample'. the final program will have to loop through all electoral districts.

```python
import pandas as pd
import json
import urllib.request
import numpy as np
```

```python
#This program reads in the statscan data for one specific district. Then it creates a df with a selection of the required information
#Obviously we will need to establish multiple calls, for each electoral district, and probably each Province and Canada as a whole.
#Note that electoral district calls have a 2013 year (because that is the time when they were re-established)
#Choose topic = 0 because it provides the general information
print("start")
#Canada data call
#url = "https://www12.statcan.gc.ca/rest/census-recensement/CR2016Geo.json?lang=E&geos=FED&cpt=01"
url = "https://www12.statcan.gc.ca/rest/census-recensement/CPR2016.json?lang=E&dguid=2016A000011124&topic=0&notes=0"
data = urllib.request.urlopen(url).read()

#Statscan data has a 'head' that needs to be removed, then the json can be read properly.
data1 = data[2:]
output = json.loads(data1)
# content = json.dumps(output, indent = 4, sort_keys=True) - only needed if we want to check transfer
# print(content)
```

```python
#Getting the information for creation of the dataframe
    
info_surv = output.get("DATA")
column_names = output.get("COLUMNS")

column_names

#info_surv
```

```python
#Creating the dataframe

info_surv2 = np.transpose(info_surv)

data_dict = {}

for i in range (0, len(column_names)):
    data_dict[column_names[i]] = info_surv2[i]
 
surv_df = pd.DataFrame.from_dict(data_dict)

surv_df.rename(columns={'PROV_TERR_NAME_NOM':'PROV_NAME', 'GEO_NAME_NOM':'NAME', 'TEXT_NAME_NOM':'TEXT','T_DATA_DONNEE':'DATA'}, inplace=True)
surv_df.head(10)
```

```python
#Simplifying the dataframe, also need complete display for analysis
pd.set_option('display.max_rows', 1000)
print(list(surv_df))

#Simplifying Dataframe - Really unnecessary, but makes work easier
surv2_df = surv_df[['PROV_NAME', 'GEO_UID', 'GEO_ID', 'NAME', 'TEXT', 'DATA']]
surv2_df = surv2_df.reset_index()
surv2_df.drop(['index'], axis=1, inplace=True)

# Taking several cuts to find elements in the dataframe
surv2_df.head()
```

```python
#Array with required information from the dataframe
column_names = ['PROVINCE', 'GEO_UID', 'GEO_ID', 'NAME']
distr_data = [surv2_df.iloc[0, 0], surv2_df.iloc[0, 1], surv2_df.iloc[0, 2], surv2_df.iloc[0, 3]]

req_info = [0, 5, 38, 57, 114, 115, 116, 673, 859, 1140, 1141]
for req in req_info:
    c_name = surv2_df.iloc[req, 4]
    d_data = surv2_df.iloc[req, 5]
    column_names.append(c_name)
    distr_data.append(d_data)
    print(c_name, d_data)

print(column_names, distr_data)

```

```python
#Setting up the new dataframe - this is just an example, we need to create the dataframe with all the data

df2 = pd.DataFrame(np.array([distr_data]), columns=[column_names])
df2.drop(['GEO_UID', 'NAME'], axis=1, inplace=True)
df2.head()
```
